```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request
from fake_useragent import UserAgent

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless') #내부 창을 띄울 수 없으므로 설정
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
# 아래 셀에서 Mozilla ~ 로 시작하는 부분을 자신의 컴퓨터에 맞게 바꿔주어야 함.
# http://www.useragentstring.com/
# 위 링크 들어가서 확인하기
ua = UserAgent(use_cache_server=True)
userAgent = ua.random
print(userAgent)
chrome_options.add_argument(f'user-agent={userAgent}')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.implicitly_wait(5)
headers = {
    'User-Agent': ua.random,
}


title_news = []
content_news = []
company_news = []
url_base='https://linkareer.com/cover-letter/search' 

#https://linkareer.com/cover-letter/search?page=20&tab=all
for p in range(1, 10, 1): #571
  url = url_base  + '?page=' + str(p)+'&tab=all'
  res = requests.get(url, headers = headers)
  soup = BeautifulSoup(res.text, 'html.parser')

  links = []

  news_links = soup.select("div.MuiBox-root")# selects all .news_item's
  for i in news_links:
    for tag in i('a'):
      if 'cover-letter' in tag.get('href') and 'search' not in tag.get('href'):
        links.append(tag.get('href'))

  

  





```

```python
from selenium.webdriver.common.by import By
df=pd.DataFrame(columns=['link','txt'])
for i in range(584,1000):
  url='https://linkareer.com'+links[i]
  print(i, '  ',url)
  chrome_options = webdriver.ChromeOptions()
  chrome_options.add_argument('--headless') #내부 창을 띄울 수 없으므로 설정
  chrome_options.add_argument('--no-sandbox')
  chrome_options.add_argument('--disable-dev-shm-usage')
  # 아래 셀에서 Mozilla ~ 로 시작하는 부분을 자신의 컴퓨터에 맞게 바꿔주어야 함.
  # http://www.useragentstring.com/
  # 위 링크 들어가서 확인하기
  ua = UserAgent(use_cache_server=True)
  userAgent = ua.random
  print(userAgent)
  chrome_options.add_argument(f'user-agent={userAgent}')
  driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
  driver.implicitly_wait(5)
  try:
    driver.get(url)
    a=driver.find_element(By.CSS_SELECTOR,"article#coverLetterContent")
    print(a)
    temp=pd.DataFrame({'link':url,'txt':a.text},index=[0])
    df=pd.concat([df,temp])
    df.to_csv("/content/drive/MyDrive/링커리어_1000.csv",encoding='utf-8-sig',index=False)
  except:
    continue

  
df
```
