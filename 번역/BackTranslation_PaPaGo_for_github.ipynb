```python
from selenium.common.exceptions import UnexpectedAlertPresentException, NoSuchElementException
```

```python
from selenium.webdriver.common.by import By
translated = []
def eng_to_kor(text, trans_lan):
  for i in tqdm(range(len(text))):
    try:
      driver.get('https://papago.naver.com/?sk='+trans_lan+'&tk=ko&hn=0&st='+text[i])##
      time.sleep(2)
      backtrans = backtrans = driver.find_element(By.XPATH,r'//*[@id="txtTarget"]').text
      print(backtrans)
      translated.append(backtrans)
    except UnexpectedAlertPresentException:
      translated.append('NO')
      print("NO")
      pass
    except NoSuchElementException:
      translated.append('NO')
      print("NO")
      pass
    except: 
      driver.get('https://papago.naver.com/?sk='+trans_lan+'&tk=ko')
      driver.find_element(By.XPATH,r'//*[@id="txtSource"]').send_keys(text[i]) 
      time.sleep(2.5) 
      backtrans = driver.find_element(By.XPATH,r'//*[@id="txtTarget"]').text 
      print(backtrans)
      translated.append(backtrans)
```

# data 위의 것 다시 실행

```python
eng_to_kor(data['TEXT'], 'en')
data['eng_to_kor'] = translated
```
