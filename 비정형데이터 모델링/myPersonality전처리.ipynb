# 특수문자, 이모지 지우기

```python
import pandas as pd
df=pd.read_excel("/content/drive/Shareddrives/22-2 캡스톤 프로젝트/데이터/myPersonality_groupby.xlsx")
df.head()
```

```python
# 이모지 지우기
import re
emoji=[':>',':D','3>',':)',':(','):','(:','=>','=D','XO',':-)','<3',':]','=)','lol','XD',':-(','> <']
exclusions = '|'.join(emoji)
df['KOR_TEXT']=df['KOR_TEXT'].apply(lambda x: re.sub(re.escape(exclusions),'',x))
df.head()
```

```python
df.to_csv("/content/drive/Shareddrives/22-2 캡스톤 프로젝트/데이터/preprocessed_personality.csv",index=False,encoding='utf-8-sig')
```

```python
import re
 
def clean_text(inputString):
  text_rmv = re.sub('[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', ' ', inputString)
  return text_rmv
df['KOR_TEXT']=df['KOR_TEXT'].apply(lambda x: clean_text(x))
```
